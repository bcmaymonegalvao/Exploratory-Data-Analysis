# Exploratory Data Analysis

The following notebook contains a series of Python operations aimed at performing exploratory data analysis on the **Titanic Survivors Dataset.**

This script aims to answer the following question:

- **Would it be possible to determine whether a Titanic passenger survived?**

**Dataset:** https://www.kaggle.com/competitions/titanic/data

The dataset columns are:

- **Passenger ID:** Unique number for each passenger.
- **Survived:** Did the passenger survive? (0 = No, 1 = Yes).
- **Pclass:** Passenger class (1 = First class, 2 = Second class, 3 = Third class).
- **Name:** Passenger name.
- **Sex:** Passenger gender.
- **Age:** Passenger age.
- **SibSp:** Number of siblings/spouse aboard the Titanic.
- **Parch:** Number of parents/children aboard the Titanic.
- **Ticket:** Ticket number.
- **Fare:** Price of the ticket.
- **Cabin:** Cabin number.
- **Embarked:** Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton).

### Visualizing graphically

```python
# Importing matplotlib
import matplotlib.pyplot as plt
```

```python
# Visualizing the above information in graphical mode
fig, ax = plt.subplots()

ax.bar(base.nunique().sort_values(ascending = False).index, base.nunique().sort_values(ascending = False).values, color = '#6bbc6b')

ax.tick_params(axis = 'x', labelrotation = 75)

plt.show()
```

```python
# Verifying the fare histogram
fig, ax = plt.subplots()

ax.hist(base.Fare, bins=20, color = '#6bbc6b')

plt.show()
```

```python
#  Verifying the fare histogram for below 100 dolar values only
fig, ax = plt.subplots()

ax.hist(base[base.Fare < 100].Fare, bins=20, color = '#6bbc6b')

plt.show()
```

### Boxplot graphic

```python
# Verifying for the Fare Column
fig, ax = plt.subplots()
ax.boxplot(base.Fare)

plt.show()
```

```python
# Acessing the .describe() values
base.Fare.describe()
```

```python
# Determining the interquartile
Q1 = base.Fare.describe()['25%']
Q3 = base.Fare.describe()['75%']
interquartile = Q3 - Q1

# Calculating the max_value
max_value = Q3 + 1.5*interquartile
```

```python
max_value
```

```python
interquartile
```

```python
# Now, filtering values above the max_value
base[base.Fare > max_value]
```

**Interpreting these values**

```python
# Couting the number of the lines
base.shape[0]/4
```

```python
# Checking the number of records in each interval (<Q1, between Q1 & Q2, between Q2 & Q3,>Q3)
Q2 = base.Fare.describe()['50%']
base[base.Fare < Q2]
```

```python
# Checking for the Sibsp and Parch columns
fig, ax = plt.subplots(ncols=2, figsize=(12,8))

ax[0].boxplot(base.Fare)
ax[1].boxplot([base.SibSp, base.Parch])

plt.show()
```
